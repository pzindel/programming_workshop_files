```python
plt.plot(data)
plt.grid()
plt.ylabel("Position, y(m)")
plt.xlabel("Time, t(s)")
plt.title("Noisy Sine Wave")
plt.xlim(0, 400)
plt.show()
```

```python
data_fft = np.abs(np.fft.fft(data))

plt.plot(data_fft[:len(data_fft)//2])
plt.grid()
plt.ylabel("Amplitude")
plt.xlabel("Frequency, f(Hz)")
plt.title("FFT of Noisy Sine Wave")
plt.show()
```
