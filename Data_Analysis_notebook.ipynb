### Analysis

```python
fig, ax = plt.subplots(figsize=(12,6))

for pr in covid["prname"].unique():
    ax.plot(covid[covid["prname"] == pr]["date"],
           covid[covid["prname"] == pr]["numconf"],
           label=pr)
    
ax.legend()
ax.grid()
ax.set_xlabel("Date")
ax.set_ylabel("Number of confirmed cases")
ax.set_title("Total number of confirmed cases in Canada")
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,20)
y = np.exp(x)

# linear plot
plt.figure()
plt.plot(x,y)
plt.title("Linear Plot")
plt.show()

# log scale plot
plt.figure()
plt.plot(x,y)
plt.title("Log scale plot")
plt.yscale("log")
plt.show()
```

```python
for pr in covid["prname"].unique():
    covid[covid["prname"] == pr].plot(x="date",
                                      y="numconf",
                                      grid=True,
                                      logy=True,
                                      title="Total number of confirmed cases in "+pr+" (log scale)")
```

### Conclusion/Interpretation
