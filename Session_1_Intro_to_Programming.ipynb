### What is a computer program?
- a **set of instructions** that describe to the computer how to mange and manipulate data to solve a problem
- a program typical implements of a series of algorithms each of which is a smaller set of instructions that addresses a smaller, more particular part of the whole problem
- programs are also referred to as software, particularly when programs and data is bundled into a package and distributed for users to utilize

### Why do we want to program computers?
- modern computers are capable of storing and managing an incredible amount of information but without a program telling them how to handle the data and what to do they're no more useful than doorstop
- through programming we can **solve complex problems and automate laborious tasks** that allow us to make our lives easier and do things that would otherwise not be possible

### How can we program computers?
- programs are written in **programming languages** which use a fixed set of keywords and puncuation which can be translated in a way that computers understand the instructions we give them
- the translated language that computers understand is called **machine language** and the translated program is referred as **machine code**
- programs are translated into machine code through the use of a **compiler** or **interpreter** (depending on the programming language)
- there are many different languages which use different keywords and puncuation and each one was designed for a specific purpose
- examples of most commonly used programming languages: Java, C#, C/C++, Python, Javascript, ...
- we'll be using **Python**; a language that was designed to be simple and easy to understand, and has a large community of developers providing open-source software that can be used to learn and create with

To help with your understanding of conecpts it can sometimes be useful to think of them in terms of different situations. In this case a good analogy would be cooking recipes! Recipes have been around for thousands of years and in very much the same way are a set of instructions for a task: 
- a chef [*the programmer*] writes down a recipe [*the program*] which includes certain keywords such as mince, whisk, mix, etc [*the programming language*]
- the recipe [*the program*] details on how to make the meal, complete with ingredients [*input data and variables*] and serving instructions [*output data*]
- a person following the recipe [*the computer*] follows the instructions to make what the chef intended

# Let's see some examples and demos!

Before we look at some examples of programs wirtten in Python we need to mention a useful tool in every programmers toolbox that should be used before starting to write any program or part of a program. That tool is **Pseudocode**:
- similar to a writing assignment, pseudocode is a draft copy of your program written in plain english (or any language you prefer)
- the purpose of pseudocode is to provide an outline of the structure and flow of your program without having to worry about the correct implementation using in the programming language you're using (i.e. focus on the what and not the how)

Here's an example for the pseudocode for a basic calculator:

$
\text{Get the first number from the user} \\
\text{Get the second number from the user} \\
\text{Get the operation from the user} \\
$

$
\text{If the operation is addition} \\
    \hspace{10mm}\text{Print the sum} \\
\text{If the operation is substraction} \\
    \hspace{10mm}\text{Print the difference} \\
\text{If the operation is multiplication} \\
    \hspace{10mm}\text{Print the product} \\
\text{If the operation is division} \\
    \hspace{10mm}\text{Print the quotient} \\
$

Writing out the pseudocode first lets us get a feel if we are missing something or there is a flaw in our logic. For example if we had forgotten to ask the user for an *operation* in line 3 than line 4 wouldn't be valid. Additionally we can see the full requirements of the *operation* variables so that when we implement it we can provide the user with a list of valid operations.

Note that we could replace any statement in our pseudocode with another statement of similar meaning (e.g. changing $\text{Get the first number from the user}$ to $\text{Ask the user for a number}$) without it altering the effectiveness of the pseudocode. The same isn't true for actual code; although we can do tasks in a number of ways we need to use the defined keywords and synatx to write our code.

Here's an example of the above pseudocode translated to Python code that the computer can handle:

```python
# Get the user's input
number_1 = int(input("Enter the first number: "))
number_2 = int(input("Enter the second number: "))
operation = input("Enter the operation (use one of + - * /): ")

# Print the solution
if operation == '+':
    print("The sum of the numbers is:", number_1+number_2)
if operation == '-':
    print("The difference of the numbers is:", number_1-number_2)
if operation == '*':
    print("The product of the numbers is:", number_1*number_2)
if operation == '/':
    print("The quotient of thenumbers is:", number_1/number_2)
```

With just a few lines of code we have a complete program that is as good as most basic calculators. You'll notice that the Python code looks very similar to the pseudocode which is why Python is considered an ideal language for beginners and those who want to quickly test programs without having to worry about every little detail of programming.

This example already shows us four very important components of a program: 
- input [*the numbers the users gives the program*]
- output [*the messages printed to the screen*]
- conditional decisions [*the if statements that determine what to do based on the operator*]
- arithmetic operation [*which manages the data stored by the program to solve a problem*]

We'll be covering these and several other components in future sessions.

### Cryptography & Security:

Cryptography is practice that has been found to date back to Ancient Egyptian times. It has been used throughout the ages to protect the secrecy of important correspondents especially those with of military significance. It is still a major area of focus in modern computers and software development. Encryption is a vital part of information transmission over the internet and ensuring that people can go privately go about their business.

The basic method of encryption is to create a plain text message and alter it using a specific method or key. The resulting message will then appear to be nonesense until someone comes along and decrypts the garbled messages using a reversed method. Here's an example in Python that uses the $\href{https://en.wikipedia.org/wiki/ROT13}{\text{ROT13 substitution cipher}}$:

```python
# Starting with a plain text message
plain_text = "we ride at dawn"
print("Original message: ", plain_text)

# We encrypt it using ROT13, rotating each letter by 13 down the alphabet
cipher_text = "".join([c if (c == " ") else chr(ord(c)+13) if (ord(c)+13) < 123 else chr((ord(c)+13)-26) for c in plain_text])
print("Encrypted message: ", cipher_text)

# We decrypt the message using the reverse process to recover the original message
decrypted_text = "".join([c if (c == " ") else chr(ord(c)-13) if (ord(c)-13) > 96 else chr((ord(c)-13)+26) for c in cipher_text])
print("Decrypted message: ", decrypted_text)
```

### Batch Operations & Task Automation:

One of the simplest and best applications of coding is for automating repetitive and boring tasks. With just a few basic lines that any beginner can understand you can potential do a task hundreds of times. 

Here's an example of a program thats goes through all the images in a folder and resizes the images to half and then saves a new copy the image with the resolution size in the filename. Notice how the program doesn't specify how many images it just goes through the entire folder which means you could resize a thousand photos or just two without having to change the program.

```python
# Get a list of file names in the images folder
from os import listdir
files = listdir("session_1_images/")
print(files)

# Process each file in the directory
for file in files:
    # Open the image
    filename = "session_1_images/" + file
    image = Image.open(filename)
    # Resize the image
    image = image.resize((image.width//2, image.height//2))
    # Find the new filename and save the image
    filename = filename[:-4] + "_" + str(image.width) + "x" + str(image.height) + filename[-4:]
    image.save(filename)
```
