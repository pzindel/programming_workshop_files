# What is data analysis?

- the process of cleaning, transforming, and modeling data so that we can discover useful information about the data that allows us to make knowledgable conclusions and decisions
- often used in business environments to aid the direction and growth of operations
- used in academia and science to confirm hypothesis and to demonstrate the validity and practicality of theories
- involves a number of subprocesses including: *collection*, *cleaning*, *analysis*, *visualization*, *interpretation*

### The process

Collection
- the subprocess of gathering and measuring information on the interested topic
- may involve collecting first-hand data by performing experiements or collecting second-hand data by scrapping relevant available databases
- when using Python data is often stored in local CSV, JSON, or text file, SQL, APIs
- we can also use SQL to connect to a database and retrieve information, or use an APIs provided by companies and developers to access data
- web scraping is another option commonly used to collect large amounts of data from webpages directly

***
Cleaning
- the subprocess of removing or modifying data that is incorrect, incomplete, irrelevant, duplicated, or improperly formatted to prepare the dataset for analysis
- may also involve combining data from multiple sources/files and properly formating and reshaping to datasets to a unified structure
- the aim of this process is to have all entries using the same format and remove any sources of incorrectness that may hinder the analysis and provide inaccuracies in the results
- the data may also need checking for statistical imblanaces that will effect the analysis

***
Analysis
- the subprocess of exploring the cleaned data and analyzing to discover information related to your objective
- this process is heavily dependent on your objective which may range from observing a trend that confirms a hypothesis or correlation to clustering data points and detecting anomlies
- may also involve building statistical models and creating visualizations like graphs, maps, and charts that accurately display the data and make it easy to understand the patterns you want to show

***
Interpretation/Application
- once the data has been fully analyzed and visualized you will want to interpret the results from the analysis to address the original objective and any other interesting things you may have found
- this subprocess may include creating live dashboards and reports that validate the direction of future experiments or business directions using the analysis

***

The libraries can be imported to our environment using their lowercase names. For Matplotlib we'll only be importing a subsection of the library containing the graphing functions we want to use by appending *.pyplot*. When importing the libraries we typically assign the shortforms *np* for NumPy and *plt* for Matplotlib.pyplot for easier referencing.

```python
import numpy as np
import matplotlib.pyplot as plt

# This following line needs to be included to ensure Juptyer NoteBook prints our plots inline with our code
%matplotlib inline
```

We can now open a data file using NumPy's *loadtxt()* function which will automatically enter the data into an array structure and return it to us. 

The dataset we'll start with is a simple 1D set containing only 2000 float values which correspond to a noisy sine wave signal. In some cases you may need to look at the formating of the dataset before loading it into Python.

Note: NumPy is great at loading purely numerical datasets but not mixed types. If the dataset contains a header of strings and columns of strings you will need to use the *skiprows* and *usecols* parameters to ignore those entries.

```python
data = np.loadtxt("session_3_data/noisy_sine_wave.csv", delimiter=',')
print(type(data))
print(data)
```

We can now explore the dataset using builtin methods and variables such as *dtype* and *shape*.

```python
print(data.dtype)
print(data.shape)
```

NumPy arrays can be indexed and sliced into sections the same way as done with Python lists.

```python
print("first element: ", data[0])
print("first ten elements: ", data[:10])
```

As mentioned, NumPy contains many functions that can be applied to these array and each function has a set of parameters that you can use to tweak the results.

```python
print("the largest value in the set is: ", np.max(data))
print("the smallest value in the set is: ", np.min(data))
print("the average value is: ", np.mean(data))
```

Since our data is 1D there's no ambiguity and no extra parameters are required. However consider the case where we want to find the average value of a 2D array; this can be done several different ways depending on what average we want:

e.g.

```python
x = np.array([[1,2],
              [3,4]])
print(x)

# average of all values (results in 1 value, namely (1+2+3+4)/4 = 2.5)
print("\n average of all values in matrix: ", np.mean(x))

# average of columns (results in 2 values, one for each column e.g. (1+3)/2 = 2 and (2+4)/2 = 3)
print("\n average down each column in matrix: ", np.mean(x, axis=0))

# average of rows (results in 2 values, one for each row e.g. (1+2)/2 = 1.5 and (3+4)/2 = 3.5)
print("\n average across each row in matrix: ", np.mean(x, axis=1))
```

***

Now let's use Matplotlib to visualize our data. Calling the *plot()* function on our 1D data will give a ready graph of the data with all default settings.

```python
plt.plot(data)
```

The y-axis represents the numerical value of each data point (e.g. the first point will be near -5 since the first element of the array has a value of -5.04) and the x-axis represents the number of data points.

This graph on its own is not very helpful; anyone who looks at the plot will not know what the data represents or the scale and description of the axes. We can improve our graph by calling supplementary functions that allow us to make edits. Each function has a set of parameters you can add to alter the appearence as well so be sure to call *help()* to explore each option.

```python
plt.plot(data)                      # create a graph of the data
plt.grid()                          # apply a grid to the background
plt.ylabel("Position, y(m)")        # attach a label to the y-axis
plt.xlabel("Time, t(s)")            # attach a label to the x-axis
plt.title("Noisy Sine Wave")        # attach a title to the graph
plt.xlim(100, 400)                  # adjust the graph to only show from 100 to 400 on the x-axis
plt.show()                          # display the graph
```

Since our data is of a sine wave we know that is periodic and we may be interested to know what the main frequencies are however the noise make this difficult in the current form. NumPy includes many mathematical tools that can come in handy when analyzing dataset and in our case we can use a Fourier transformation to convert our time series (i.e. time on the x-axis) to a frequency spectrum (i.e. frequency, or inverse of time, on the x-axis). 

NumPy's Fourier transform functions are stored in the *fft* module (short for *fast fourier transform*). Passing our data to the *fft.fft()* function will return the frequency spectrum. Since the returned result contains negative values and is symmetrical we'll apply the absolute function and take the first half of the results:

```python
data_fft = np.abs(np.fft.fft(data))

plt.plot(data_fft[:len(data_fft)//2])
plt.grid()
plt.ylabel("Amplitude")
plt.xlabel("Frequency, f(Hz)")
plt.title("FFT of Noisy Sine Wave")
plt.show()
```

Our frequency spectrum shows a much clearer picture of the sine wave that our intial data. We can see there are 3 primary frequencies that make up the wave with the main one around 300Hz. If we wanted to model the sine waveform without the noise present we could extract the exact frequencies and recreate it using the mathematical formula. Alternatively we could also apply a filter at this stage and apply the inverse Fourier transformation to get back a noise reduced version of our intial data.

- provides high-level data structures and functions to easily work with tabular and structured data
- DataFrame: tabular, column-oriented data structure with row and column labels
- Series: 1D labeled array object
- mixes high-performance, array-computing of NumPy with flexible data manipulation of spreadsheets
- provides sophisticated indexing to make it easy to reshape, slice and dice, perform aggregations, and select subsets of data
- flexible handling of missing data
- merge operations
***

We can start by importing the library using the keyword *pandas*. A common shortform used for the library is *pd*.

```python
import pandas as pd
```

We'll now open a new dataset using Pandas which will automatically put the entries into a Pandas DataFrame structure (which we can verify by calling the *type()* function on our variable.

```python
data = pd.read_csv("session_3_data/airline-safety.csv")
type(data)
```

The data we're looking at (which can be found here: https://github.com/fivethirtyeight/data/tree/master/airline-safety) contains a collection of airlines and the number of incidents, fatal accidents, and fatalities each had from 1985-1999 and from 2000-2014.

We can print a copy of the dataframe by calling the variable to see the overall format and contents.

```python
data
```

We can see that unlike NumPy, with Pandas we can open a data set with mixed data types and Pandas will take care of sorting the data into the corresponding variable type. Pandas also automatically creates a header from the first row and provides indicies for each entry. If our data file were to already contain indicies or missing a header we would need to specify that when calling *read_csv()*.

We can see the date type used for each column by calling the *.dtypes* attribute:

```python
data.dtypes
```

Pandas dataframes have many methods and attributes that we can call to easily explore the contents of our dataset.

- *.shape* allows us to see the number of columns and rows in our dataframe
- *.columns* prints out a list of column names
- *.head()* prints the first 5 entries in the dataframe
- *.tail()* prints the last 5 entries int the dataframe
- *.describe()* prints statistical information for entries in numerical columns

```python
print(data.shape)
print(data.columns)
data.describe()
```

You can also combine methods for specific tasks. For example we can call *.notna()* which will convert all entries' values into a boolean which is false when the data for that entry is missing. On its own this method isn't very useful since if you wanted to check for missing entries you'd still have to search the entire dataset. However we can append *.describe()* to *.notna()* to provide a breakdown of how many entries are missing in each column.

```python
data.notna().describe()
```

- *count* tells us the number of entries in each column
- *unique* tells us the number of different values in each column (since we called *.notna()* this value will be 1 or 2 depending if any entries are *false*)
- *top* tells us the unique value that comes up in the most number of entries
- *freq* tells us the number of time *top* value appears in the column

From this we can tell that our dataset doesn't have any blank values that we need to remove before analyzing the data.

Similar to Python lists we can pick out specific columns from the dataframe using indexing, however since our columns have header names we use the corresponding names instead of integers.

```python
data["airline"]
```

And we can also use a list of indecies to select several columns' data:

```python
data[["airline", "avail_seat_km_per_week"]]
```

***

We can plot data from our dataframe similarly to when we use Matplotlib. Pandas has a builtin plotting method that we can call on dataframes and series data structures. Any specifications we have to form our plot can be specified in the parameters when we call the method.

Since our data has multiple dimensions (i.e. columns) we need to select which ones we would like to plot. For example if we wanted to see which airlines have the highest passenger carrying capacity we would select the *airline* and *avail_seat_km_per_week* (we'll use *ASK* for short) columns and create a box plot as shown below. (Note: *avail_seat_km_per_week* is a measure of the weekly carry capacity found by multiplying the number of seats by the number of km flown per flight and adding all flights over the course of a week.)

```python
axes = data[["airline", "avail_seat_km_per_week"]].plot(x="airline",
                                                       y="avail_seat_km_per_week",
                                                       kind="bar",
                                                       figsize=(12,5),
                                                       grid=True,
                                                       title="Airlines' Weekly Availabe Seat Km (ASK)")
axes.set_ylabel("ASK per week (billions)")
```

This plot is a great way of seeing the imbalance in our dataset. A handfull of airline have a much higher *ASK* than others which means that they might be more prone to accidents (since they spend serval times as much time in the sky) or any accidents that occur will effect a larger number of people. These sorts of imbalances need to be taken into account when comparing data from different groups or categories.


```python
axes = data[["airline", "incidents_00_14", "fatal_accidents_00_14"]].plot(x="airline",
                                                                          kind="bar",
                                                                          figsize=(12,5),
                                                                          grid=True,
                                                                          title="Airlines' Number of Incidents and Accidents 2000-2014")
axes.set_ylabel("Count")
```

Pandas also allows us to manipulate and create new columns. We can normalize the number of fatalities by dividing it by the ASK value to get a rate that we can compare airlines with. We'll assign the normalized values to a new column named *fatalities_00_14_norm*:

```python
data["fatalities_00_14_norm"] = data["fatalities_00_14"] / data["avail_seat_km_per_week"]*1e9
data["fatalities_85_99_norm"] = data["fatalities_85_99"] / data["avail_seat_km_per_week"]*1e9
```

We can also sort the dataframe by values in a columns. By default our dataset came sorted alphabetically based on airline. When plotting it may be useful to sort the data according to a specific column since the data is plotted in order. This allows you to add more information to a 2D graph and compare values more easily.

```python
data.sort_values("avail_seat_km_per_week", ascending=False)
```

```python
axes = data.sort_values("avail_seat_km_per_week", 
                        ascending=False)[["airline",
                                          "fatalities_85_99_norm",
                                          "fatalities_00_14_norm"]].plot(x="airline",
                                                                         kind="bar",
                                                                         figsize=(12,5),
                                                                         grid=True,
                                                                         title="Airlines' Number of Fatalities 1985-2014")
axes.set_ylabel("Normalized Fataility Rate")
```

Thus by normalizing and sorting our data before plotting we see how airlines might stack up against each other. On the left we have airlines with higher ASK values but moderately low fatality rates opposed to some smaller airlines with lower ASK.

# Example

Let's now take a look at a COVID-19 dataset provided by the government and go through the steps you would take to analyze the data.

### Collection

Given the global scale and ongoing nature of the COVID-19 pandemic there is tons of data from various souces that can be analyzed for a number of reasons. If you were taking a comprehensive look at all the data you might consider reviewing data from a variety of sources especially if there's conflicting reports.

For our purposes we'll restrict ourselves to a single dataset on COVID-19 stats in Canada. For this reason we'll retrieve our data directly from the government site (https://open.canada.ca/data/) and assume it to be the most accurate reporting on the pandemic within Canada.

Searching the government database for the keyword "covid" provides many results on different related data. The one we'll look at is the Public Health Infobase (https://open.canada.ca/data/en/dataset/261c32ab-4cfd-4f81-9dea-7b64065690dc).

```python
# Import any required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# This following line needs to be included to ensure Juptyer NoteBook prints our plots inline with our code
%matplotlib inline

# Open the dataset into a pd.DataFrame
data = pd.read_csv("session_3_data/covid19.csv")
```

### Cleaning

Now that we have our data we need to explore it and see what is present and what info is missing. If entries are duplicated, incomplete or incorrect we need to find out now so that we can possible correct them, remove them, or at the very least take them into account when doing our analysis.

```python
data.dtypes
```

Checking the names and data types for each column we see that all but 3 of them are numerical. Two of the non-numerical columns are province names (one English and one French); since they provide the same data and we'll only use the Enlgish names we can remove the *prnameFR* column before analysis to simplify the dataframe.

The 3rd non-numerical is *date* which gives us the day corresponding to the entry data. The current format seems to be string (which shows as *object* when we call *.dtypes*) which we'll want to convert to a datetime variable.

```python
# Remove duplicate column
del data["prnameFR"]

# Print the type of the date value of the first entry
print(type(data["date"][0]))

# Reassign the date using a standard format and datetime variable
data["date"] = pd.to_datetime(data["date"], format="%d-%m-%Y", infer_datetime_format=True)
data.dtypes
```

We also know that *prname* and *pruid* should have a limited number of unique values that correspond to provinces. We can check that's the case by calling the method *.unique()*:

```python
data.prname.unique()
```

```python
data.pruid.unique()
```

We can assume that each *pruid* is unique to its corresponding region in *prname*. Since we won't be using the *pruid* column we don't need to check but if you wanted to you could run *.unique()* on a subset of entries for each region and ensure there's only one uid. E.g.

```python
data[data["prname"] == "Ontario"]["pruid"].unique()
```

Next we'll check for missing values in our entries:

```python
data.isna().describe()
```

```python
data.head()
```

```python
data.tail()
```

which shows that most columns are missing some values. Notabley *numrecover*, *percentrecover*, and a number of daily stats (deaths, tested, recovered) are missing quite a few, presumabley from there being no reported recoveries at the start of the pandemic outbreak and less rigorous stat tracking. We could choose to convert the values from NaN to 0 but for now we'll just filter out missing values when we want to look at those columns.

```python
data.describe()
```

### Analysis

Now that we know what's in our dataset we can start to explore and graph different relations to see how the virus outbreak has progressed since the start of February.

```python
# Check for the regions in our dataset
data['prname'].unique()
```

```python
# Create a Matplotlib figure so we can plot each region
fig, ax = plt.subplots(figsize=(12,5))

# Interate the regions and plot
for pr in data['prname'].unique():
    ax.plot(data[data["prname"] == pr]["date"], data[data["prname"] == pr]["numconf"], label=pr)

# Format our plot
ax.legend()
ax.grid()
ax.set_xlabel("Date")
ax.set_ylabel("Number of Confirmed Cases")
ax.set_title("Total Number of Confirmed Cases in Canada")
```

Given how the virus spreads from person to person, the growth if let uncheck is exponential. Instead our plot above shows linear-like growth. Switching to a logarithmic scale can give us a better sense of this since exponentional growth will appear linear on a logarithmic plot. E.g.

```python
# Set points for a exponential line
x = np.arange(0, 20)
y = np.exp(x)

# Plot on linear scale
plt.figure()
plt.plot(x,y)
plt.title("Linear Plot")
plt.show()

# Plot on log scale
plt.figure()
plt.plot(x,y)
plt.title("Log Plot")
plt.yscale("log")
plt.show()
```

So let's plot each total number of confirmed cases for each region on a separate logarithmic plot:

```python
for pr in data['prname'].unique():
    data[data["prname"] == pr].plot(x="date", 
                                    y="numconf", 
                                    grid=True,
                                    logy=True,
                                    title="Total Number of Confirmed Cases in "+pr+" as of May 21, 2020")
```

Having separate plots allows us to look at the stats of each province more closely without massive values making smaller values unreadable. Since the curves in the log plot aren't completely linear we see that effects to slow the spread are working quite well.

We can also take a look at the number of active cases in each region to see whether the number seems to be declining. We can find the number of active cases by removing the recovered and death values from the total.

```python
for region in data['prname'].unique():
    subset = data[data["prname"] == region].fillna(0)
    
    # Number of active cases = number of confirmed cases - (number of deaths + number of recovered)
    x = subset["date"]
    y = subset["numconf"] - subset["numdeaths"] - subset["numrecover"]
    fig, ax = plt.subplots(figsize=(12,5))
    
    ax.plot(x, y)
    
    ax.grid()
    ax.set_xlabel("Date")
    ax.set_ylabel("Number of Active Cases")
    ax.set_title("Active Cases in "+region)
```

As the pandemic progresses we expect the number of active cases to increase from zero, reach a peak, and return to near zero (tracing out a bell shape). 

We see that most provinces are on the downward slope of the bell curve. Interestingly several plots have more than one peak which happens when safety precautions that slow the spread are lifted too early or mass gathering happen.

We also see that Quebec's trend has not reach its peak yet and since they have the most number of active cases by far, the trend for all of Canada is also still upwards.

Lastly we'll take a look a the number of new cases vs the total number of cases on a weekly basis. Note that the entries for the last week of our data may not be complete so we'll omit it.

```python
region = "Canada"
# Select (entire Canada region and exclude incomplete week) and group (by the week) the entries
subset = data[(data["prname"] == region) & (data["date"].dt.week < 22)].groupby(data["date"].dt.week)
   
# Aggregate the values
y = subset["numtoday"].sum()
x = subset["numconf"].max()
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(x, y)
  
# Format the plot
ax.set_yscale('log')
ax.set_xscale('log')
ax.legend()
ax.grid()
ax.set_xlabel("Total Number of Confirmed Cases")
ax.set_ylabel("Number of New Cases (weekly)")
ax.set_title("Weekly New Case Growth in "+region)
```

### Interpretation

From our plots we can see the spread of the virus over the last few months and how quarantine efforts have managed to greatly reduce uncontrolled spread. It's also clear that majority of provinces are seeing an overall decline in active cases but easing up on isolation and distancing measures could result in a resurrgence in cases.
